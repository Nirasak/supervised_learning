```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
```

## Dataset infos

For this exercice we are gonna pick the Iris dataset from sklearn which is often used for exercices in machine learning.
The Iris dataset contains 150 samples about the sepal and petal length and width of three different species of iris flowers: setosa, versicolor, and virginica. The goal is to predict the species of an iris flower based on these four features.

First of all let's take a look at the dataset and its correlation matrix

```python
data = load_iris()

df = pd.DataFrame(data=np.c_[data['data'], data['target']],
                  columns=data['feature_names'] + ['target'])

print(np.shape(df))
print(df.describe())

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".3f")
plt.title("Correlation Matrix")
plt.show()
```

This correlation matrix provides insights into the relationships between different features and the target variable in the iris dataset. Here are some observations:

- Sepal Length vs. Petal Length and Width: Sepal length exhibits a strong positive correlation with petal length (0.87) and petal width (0.82). This suggests that as sepal length increases, so does the length and width of the petal.
- Sepal Width vs. Petal Length and Width: Sepal width shows a moderate negative correlation with petal length (-0.43) and petal width (-0.37). This implies that as sepal width increases, petal length and width tend to decrease slightly.
- Target (Species) Correlation: The target variable (species) exhibits strong positive correlations with petal length (0.95), petal width (0.96), and sepal length (0.78). This indicates that certain species may have characteristic differences in these features, making them useful for species classification.

```python
# Separate features and target variable
X = df.drop('target', axis=1)
y = df['target']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dictionary of models
classifiers = {
    "Logistic Regression": LogisticRegression(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(),
}

for name, clf in classifiers.items():
    scores = cross_val_score(clf, X_scaled, y, cv=5, scoring='accuracy')
    print(f"{name}: Accuracy: {scores.mean():.4f} (+/- {scores.std() * 2:.4f})")
```

# Results and conclusion

These results indicate the performance of three different classification algorithms on the iris dataset, evaluated using cross-validation. Here are some observations:

- Logistic Regression and Gradient Boosting: Both logistic regression and gradient boosting achieve similar accuracies, with an average accuracy of 96.00%. The slight difference in their standard deviations suggests that logistic regression might have slightly more variability in performance across different folds compared to gradient boosting.
- Support Vector Machine (SVM): SVM performs slightly better than logistic regression and gradient boosting, with an average accuracy of 96.67%. Additionally, it has the lowest standard deviation (0.0422), indicating a more consistent performance across different cross-validation folds compared to the other two algorithms.

Overall, all three algorithms demonstrate high accuracies, suggesting that they are effective at classifying iris species based on the provided features. However, SVM appears to have a slight edge in terms of average accuracy and consistency across different folds, making it a potentially preferable choice for this particular dataset.
